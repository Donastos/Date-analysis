# Задание 1

Будем парсить страницу со свежеми новостям на habr.com/ru/all/.

Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:

KEYWORDS = [‘python’, ‘парсинг’]

Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).

В итоге должен формироваться датафрейм вида: <дата> - <заголовок> - <ссылка>

```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
#определяем список хабов, которые нам интересны
KEYWORDS = ['python', 'парсинг']
```

```python
# получаем страницу с самыми свежими постами
req = requests.get('https://habr.com/ru/all/')
url = 'https://habr.com/ru/all/'
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
# извлекаем посты
posts = soup.find_all('article',class_ = 'tm-articles-list__item')
posts
```

```python
# ищем посты
for post in posts:
    
    articles_intro = post.find_all('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2')[0].span.text
    
    #     проверяем, есть ли ключевое слово в заголовке
    if any([(keyword in articles_intro) for keyword in KEYWORDS]):
        continue
    
    print(articles_intro)
```

```python
# получаем ссылки
habr_links = 'https://habr.com'

for post in posts:
    
    a_list = post.find_all('a', class_='tm-article-snippet__title-link')[0].get('href')
    
    
    all_refs = habr_links + a_list
    
    print(all_refs)
```

```python
# ищем время
date = pd.to_datetime(post.find_all('span', class_= 'tm-article-snippet__datetime-published')[0].time.get('title'))
print(date)
```

```python
# создаем пустой DataFrame() и объединям полученные данные

titles_with_keywords = pd.DataFrame()
row = {'дата': date, 'заголовок': articles_intro, 'ссылка': all_refs} 
titles_with_keywords = pd.concat([titles_with_keywords, pd.DataFrame([row])])  
titles_with_keywords
```

Написать скрипт, который будет проверять список e-mail адресов на утечку при помощи сервиса Avast Hack Ckeck. Список email-ов задаем переменной в начале кода:
EMAIL = [xxx@x.ru, yyy@y.com]

В итоге должен формироваться датафрейм со столбцами: <дата утечки> - <источник утечки> - <описание утечки>

Подсказка: сервис работает при помощи “скрытого” API. Внимательно изучите post-запросы.

```python
import requests
EMAIL = 'xxx@x.ru', 'yyy@y.com'

url_s = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
params = {
    'emailAddresses': EMAIL
}
headers = {
'Vaar-Header-App-Build-Version': '1.0.0',
'Vaar-Header-App-Product': 'hackcheck-web-avast',
'Vaar-Header-App-Product-Name': 'hackcheck-web-avast',
'Vaar-Version': '0'
}
res = requests.post(url_s, json = params, headers = headers)
res
```

```python
# формируем датафрейм
df = pd.DataFrame(res.json()['breaches']).transpose().reset_index()
df.head(20)
```

```python
# создаем требуемый
df = df[['publishDate', 'site', 'description']]
df.head(20)
```
